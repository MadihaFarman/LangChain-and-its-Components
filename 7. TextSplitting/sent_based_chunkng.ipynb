```python
import nltk
nltk.download('punkt', quiet=True)
from nltk.tokenize import sent_tokenize
document = (
"Dr. Khan reviewed the patient's chart. The diagnosis was confirmed "
"as Type 2 diabetes. Treatment includes metformin and a controlled diet. "
"Follow-up is scheduled in 6 weeks."
)
# Proper sentence tokenisation -- knows "Dr." is an abbreviation, not a sentence end
sentences = sent_tokenize(document)
print(sentences)
# ['Dr. Khan reviewed the patient's chart.', 'The diagnosis was confirmed as Type 2 diabetes.',
#'Treatment includes metformin and a controlled diet.', 'Follow-up is scheduled in 6 weeks.']
# Notice "Dr." did NOT cause an incorrect split -- a naive text.split('.') WOULD have.
def sentence_chunk(sentences, chunk_size=200, overlap_sentences=1):
    chunks = []
    current_chunk = []
    current_len = 0 
    for sent in sentences:
     if current_len + len(sent) > chunk_size and current_chunk:
         chunks.append(" ".join(current_chunk))
         # keep last N sentences for overlap into next chunk
         current_chunk = current_chunk[-overlap_sentences:]
         current_len = sum(len(s) for s in current_chunk)
     current_chunk.append(sent)
     current_len += len(sent)
    if current_chunk:
       chunks.append(" ".join(current_chunk))
    return chunks
chunks = sentence_chunk(sentences, chunk_size=80, overlap_sentences=1)
for c in chunks:
  print("---", c)
```
