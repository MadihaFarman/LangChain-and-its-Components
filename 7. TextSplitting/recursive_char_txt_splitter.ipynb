```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", ". ", " ", ""],
)
```

```python
document = """
Admission Requirements
Students applying for the BS Computer Science program must submit
the following documents: CNIC copy, academic transcripts, and two
recommendation letters.
Fee Structure
The semester fee is PKR 85,000, payable in two installments. Late
payment incurs a 5% penalty after the due date.
Recursive character splitting (LangChain’s RecursiveCharacterTextSplitter is the most widely used
implementation) fixes the "cuts mid-sentence" problem of fixed-size chunking while staying simple and fast.
The core idea: try to split on the most meaningful separator first, and only fall back to a less meaningful
separator if a chunk is still too large.
"""
```

```python
chunks = splitter.split_text(document)
```

```python
for i, chunk in enumerate(chunks):
 print(f"--- Chunk {i} ({len(chunk)} chars) ---")
 print(chunk)
 print()
```
