```python
from langchain_huggingface import HuggingFaceEndpoint
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
```

```python
# Sample documents
docs = [
    Document(page_content="LangChain makes it easy to work with LLMs."),
    Document(page_content="LangChain is used to build LLM based applications."),
    Document(page_content="Chroma is used to store and search document embeddings."),
    Document(page_content="Embeddings are vector representations of text."),
    Document(page_content="MMR helps you get diverse results when doing similarity search."),
    Document(page_content="LangChain supports Chroma, FAISS, Pinecone, and more."),
]
```

```python
# pip install faiss-cpu
```

```python
from langchain_huggingface import HuggingFaceEndpointEmbeddings

embedding_model = embedding = HuggingFaceEndpointEmbeddings(
    model= 'sentence-transformers/all-MiniLM-L6-v2',
    task= 'feature-extraction'
)


vectorstore = FAISS.from_documents(
    documents=docs,
    embedding=embedding_model
)

#This creates the FAISS vectorstore in memory only — nothing is stored on disk.
```

```python
# Enable MMR in the retriever
retriever = vectorstore.as_retriever(
    search_type="mmr",                   # <-- This enables MMR
    search_kwargs={"k": 3, "lambda_mult": 0.5}  # k = top results, lambda_mult = relevance-diversity balance
)

# mmr stands for Maximal Marginal Relevance, and it is a technique used in information retrieval to balance relevance and diversity in search results. When you perform a similarity search using MMR, it selects results that are not only relevant to the query but also diverse from each other. This helps to avoid redundancy in the results and provides a more comprehensive set of answers. In the code above, we enable MMR by setting `search_type="mmr"` and providing additional parameters in `search_kwargs` to control the number of results (`k`) and the balance between relevance and diversity (`lambda_mult`).
```

```python
query = "What is langchain?"
results = retriever.invoke(query)
```

```python
for i, doc in enumerate(results):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```
